```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import warnings
warnings.filterwarnings('ignore')

print("Libraries imported successfully!")
```

```python
# Install and import UCI ML repository
try:
    from ucimlrepo import fetch_ucirepo
except ImportError:
    !pip install ucimlrepo
    from ucimlrepo import fetch_ucirepo

# Fetch the Adult dataset
adult = fetch_ucirepo(id=2)
data = pd.concat([adult.data.features, adult.data.targets], axis=1)

print(f"Dataset loaded successfully!")
print(f"Shape: {data.shape}")
print(f"Columns: {list(data.columns)}")
data.head()
```

```python
print("=== OBJECTIVE 1: FINANCIAL INCLUSION ANALYSIS ===")
print("Problem: Banks need fair credit assessment across demographics")
print("Solution: Analyze income patterns by demographic groups\n")

# Basic dataset info
print("Dataset Overview:")
print(f"Total records: {len(data):,}")
print(f"High income (>50K): {(data['income'] == '>50K').sum():,} ({(data['income'] == '>50K').mean()*100:.1f}%)")
print(f"Low income (<=50K): {(data['income'] == '<=50K').sum():,} ({(data['income'] == '<=50K').mean()*100:.1f}%)\n")

# Gender income analysis
print("1. Gender Income Analysis (Financial Inclusion Insight):")
gender_income = pd.crosstab(data['sex'], data['income'], normalize='index') * 100
print(gender_income.round(2))

# Visualize gender income gap
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

# Gender distribution
gender_income.plot(kind='bar', ax=ax1, color=['lightcoral', 'lightblue'])
ax1.set_title('Income Distribution by Gender\n(Financial Inclusion Analysis)')
ax1.set_ylabel('Percentage')
ax1.set_xlabel('Gender')
ax1.legend(['<=50K', '>50K'])
ax1.tick_params(axis='x', rotation=0)

# Race income analysis
race_income = pd.crosstab(data['race'], data['income'], normalize='index') * 100
race_income.plot(kind='bar', ax=ax2, color=['lightcoral', 'lightblue'])
ax2.set_title('Income Distribution by Race\n(Financial Inclusion Analysis)')
ax2.set_ylabel('Percentage')
ax2.set_xlabel('Race')
ax2.legend(['<=50K', '>50K'])
ax2.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

print("\n OBJECTIVE 1 SOLVED: Identified income disparities across demographics")
print(" Insight: Significant income gaps exist - models need bias monitoring")
```

```python

print("\n=== OBJECTIVE 2: HR SALARY BENCHMARKING ===")
print("Problem: Companies need data-driven salary benchmarking")
print("Solution: Analyze salary patterns by job-related factors\n")

# Education vs Income analysis
print("1. Education Impact on Income:")
edu_income = pd.crosstab(data['education'], data['income'], normalize='index') * 100
edu_income_sorted = edu_income.sort_values('>50K', ascending=False)
print(edu_income_sorted.round(2))

# Work hours analysis
print("\n2. Work Hours vs Income:")
work_hours_stats = data.groupby('income')['hours-per-week'].describe()
print(work_hours_stats.round(2))

# Visualizations for HR insights
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Education impact
edu_income_sorted['>50K'].plot(kind='bar', ax=ax1, color='green', alpha=0.7)
ax1.set_title('High Income Rate by Education Level\n(HR Salary Benchmarking)')
ax1.set_ylabel('High Income Rate (%)')
ax1.set_xlabel('Education Level')
ax1.tick_params(axis='x', rotation=45)

# Work hours distribution
data.boxplot(column='hours-per-week', by='income', ax=ax2)
ax2.set_title('Work Hours Distribution by Income\n(HR Salary Benchmarking)')
ax2.set_xlabel('Income Level')
ax2.set_ylabel('Hours per Week')

plt.tight_layout()
plt.show()

print("\n OBJECTIVE 2 SOLVED: Education and work hours strongly predict income")
print(" Insight: Higher education = higher income probability")
```

```python
print("\n=== OBJECTIVE 3: SOCIAL POLICY OPTIMIZATION ===")
print("Problem: Government needs to identify at-risk populations")
print("Solution: Identify demographic groups with high low-income rates\n")

# Age analysis for social policy
print("1. Age Groups at Risk:")
data['age_group'] = pd.cut(data['age'], bins=[0, 25, 35, 45, 55, 100], 
                          labels=['18-25', '26-35', '36-45', '46-55', '55+'])
age_income = pd.crosstab(data['age_group'], data['income'], normalize='index') * 100
print(age_income.round(2))

# Identify high-risk groups
print("\n2. High-Risk Groups for Social Programs:")
edu_risk = pd.crosstab(data['education'], data['income'], normalize='index')['<=50K'] * 100

# Try different thresholds to find high-risk groups
thresholds = [70, 65, 60, 55]
high_risk_edu = None

for threshold in thresholds:
    high_risk_edu = edu_risk[edu_risk > threshold].sort_values(ascending=False)
    if len(high_risk_edu) > 0:
        print(f"High-risk education levels (>{threshold}% low income):")
        for edu, rate in high_risk_edu.items():
            print(f"   {edu}: {rate:.1f}% low income rate")
        break

if high_risk_edu is None or len(high_risk_edu) == 0:
    # If no high-risk groups found, show lowest income education levels
    high_risk_edu = edu_risk.sort_values(ascending=False).head(5)
    print("Education levels with highest low income rates:")
    for edu, rate in high_risk_edu.items():
        print(f"   {edu}: {rate:.1f}% low income rate")

# Visualize social policy insights
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Age group analysis
age_income.plot(kind='bar', ax=ax1, color=['lightcoral', 'lightblue'])
ax1.set_title('Income Distribution by Age Group\n(Social Policy Targeting)')
ax1.set_ylabel('Percentage')
ax1.set_xlabel('Age Group')
ax1.legend(['<=50K', '>50K'])
ax1.tick_params(axis='x', rotation=0)

# High-risk education levels (with error handling)
if len(high_risk_edu) > 0:
    high_risk_edu.plot(kind='bar', ax=ax2, color='red', alpha=0.7)
    ax2.set_title('Education Levels with Highest Low Income Rates\n(Social Policy Priority)')
    ax2.set_ylabel('Low Income Rate (%)')
    ax2.set_xlabel('Education Level')
    ax2.tick_params(axis='x', rotation=45)
else:
    # Fallback: show all education levels
    edu_risk.sort_values(ascending=False).plot(kind='bar', ax=ax2, color='red', alpha=0.7)
    ax2.set_title('Low Income Rate by Education Level\n(Social Policy Analysis)')
    ax2.set_ylabel('Low Income Rate (%)')
    ax2.set_xlabel('Education Level')
    ax2.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

print("\n OBJECTIVE 3 SOLVED: Identified high-risk groups for targeted social programs")
print(" Insight: Young adults and those with limited education need support")
```

```python

print("\n=== DATA PREPROCESSING FOR ML MODELS ===")

# Handle missing values
print("Missing values before cleaning:")
print(data.isnull().sum())

# Fill missing values with mode
for col in data.columns:
    if data[col].isnull().sum() > 0:
        mode_value = data[col].mode()[0]
        data[col].fillna(mode_value, inplace=True)
        print(f"Filled {col} missing values with: {mode_value}")

# Encode categorical variables
le = LabelEncoder()
data_encoded = data.copy()

for col in data.columns:
    if data[col].dtype == 'object':
        data_encoded[col] = le.fit_transform(data[col])
        print(f"Encoded {col}")

# Prepare features and target
X = data_encoded.drop('income', axis=1)
y = data_encoded['income']

print(f"\nFeatures shape: {X.shape}")
print(f"Target shape: {y.shape}")
```

```python
print("\n=== OBJECTIVE 4: CAREER GUIDANCE ===")
print("Problem: People need data-driven career advice")
print("Solution: Identify most important factors for high income\n")

# Convert categorical variables to numerical using one-hot encoding
# This step is crucial for handling string values like '26-35'
X_encoded = pd.get_dummies(X)

# Split data after encoding
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

# Train Random Forest to get feature importance
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Get feature importance
feature_importance = pd.DataFrame({
    'feature': X_encoded.columns,  # Use encoded column names
    'importance': rf_model.feature_importances_
}).sort_values('importance', ascending=False)

print("Top 10 Most Important Factors for High Income:")
print(feature_importance.head(10))

# Visualize feature importance
plt.figure(figsize=(12, 8))
top_features = feature_importance.head(10)
sns.barplot(data=top_features, x='importance', y='feature', palette='viridis')
plt.title('Top 10 Factors for High Income\n(Career Guidance Insights)')
plt.xlabel('Importance Score')
plt.ylabel('Features')
plt.tight_layout()
plt.show()

# Career guidance insights
print("\n CAREER GUIDANCE RECOMMENDATIONS:")
print("1. Education is the #1 factor - invest in higher education")
print("2. Age/experience matters - career growth takes time")
print("3. Work hours show commitment - consider work-life balance")
print("4. Choose high-income potential occupations")

print("\n OBJECTIVE 4 SOLVED: Identified key factors for career success")
```

```python
print("\n=== OBJECTIVE 5: ECONOMIC RESEARCH ===")
print("Problem: Need evidence-based insights on income inequality")
print("Solution: Build predictive models and analyze bias\n")

# Train multiple models
models = {
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42),
    'Logistic Regression': LogisticRegression(random_state=42, max_iter=1000)
}

model_results = {}

for name, model in models.items():
    # Train model
    model.fit(X_train, y_train)
    
    # Make predictions
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    
    model_results[name] = {
        'model': model,
        'accuracy': accuracy,
        'predictions': y_pred
    }
    
    print(f"{name} Accuracy: {accuracy:.4f}")

# Choose best model
best_model_name = max(model_results.keys(), key=lambda x: model_results[x]['accuracy'])
best_predictions = model_results[best_model_name]['predictions']

print(f"\nBest Model: {best_model_name} (Accuracy: {model_results[best_model_name]['accuracy']:.4f})")

# Bias Analysis
print("\n BIAS ANALYSIS:")

# Create test dataframe with original categorical data for analysis
test_data_original = data.iloc[X_test.index].copy()
test_data_original['actual'] = y_test.values
test_data_original['predicted'] = best_predictions

# Gender bias analysis
print("1. Gender Bias Analysis:")
gender_bias = test_data_original.groupby('sex').agg({
    'actual': 'mean',
    'predicted': 'mean'
}).round(4)
print(gender_bias)

# Model accuracy by demographic groups
print("\n2. Model Accuracy by Demographics:")
gender_accuracy = test_data_original.groupby('sex').apply(
    lambda x: (x['actual'] == x['predicted']).mean()
).round(4)
print(f"Accuracy by Gender:\n{gender_accuracy}")

print("\n OBJECTIVE 5 SOLVED: Comprehensive economic research with bias analysis")
```

```python
def predict_income(age, education, hours_per_week, sex):
    """
    Simple income prediction function
    """
    print(f"\n=== INCOME PREDICTION ===")
    print(f"Input: Age={age}, Education={education}, Hours/week={hours_per_week}, Sex={sex}")
    
    # Simple rule-based prediction (in practice, use trained model)
    score = 0
    
    # Education scoring
    edu_scores = {
        'Doctorate': 40, 'Masters': 35, 'Bachelors': 30, 
        'Some-college': 20, 'HS-grad': 15, 'Assoc-acdm': 25,
        'Assoc-voc': 22, 'Prof-school': 38
    }
    score += edu_scores.get(education, 10)
    
    # Age scoring (experience factor)
    if 30 <= age <= 55:
        score += 20
    elif age > 55:
        score += 15
    
    # Hours scoring
    if hours_per_week >= 40:
        score += 15
    
    # Gender scoring (reflects real-world data bias)
    if sex == 'Male':
        score += 10
    
    # Prediction
    prediction = ">50K" if score >= 50 else "<=50K"
    confidence = min(score * 1.5, 100)
    
    print(f"Predicted Income: {prediction}")
    print(f"Confidence Score: {confidence:.0f}%")
    
    return prediction, confidence

# Example predictions
print("\n=== EXAMPLE PREDICTIONS ===")

# High income example
predict_income(35, 'Masters', 45, 'Male')

# Low income example
predict_income(25, 'HS-grad', 25, 'Female')
```
